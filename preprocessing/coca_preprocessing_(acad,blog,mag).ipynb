```python
from bs4 import BeautifulSoup
import re
def cleaning_1(coca_data):
  sentences = []
  for i in range(1,len(coca_data)):
    coca_data_i = coca_data[i][10:] # 문서 태그 제거
    #re.sub(r"http\S+", "", coca_data_i)
    coca_data_i = re.sub('http : //\S+|https : //\S+', '', coca_data_i, flags=re.MULTILINE)
    coca_data_i = re.sub("[\(\[].*?[\)\]]", '', coca_data_i)  #https://stackoverflow.com/questions/14596884/remove-text-between-and
    cleaned_text = BeautifulSoup(coca_data_i, 'lxml').text
    splitted = nltk.tokenize.sent_tokenize(cleaned_text)
    splitted = filter(lambda x : len(x) > 10, splitted)
    splitted = filter(lambda x : "@ @ @ @ @ @ " not in x, splitted)
    sentences.extend(splitted)
    

  return sentences
```
