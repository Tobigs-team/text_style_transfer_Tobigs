```python
import pandas as pd
import re
```

## 가천대 학습자 코퍼스

```python
data = pd.read_csv("/Users/chohyowon/Desktop/Gachon Learner Corpus - Final Version.csv", error_bad_lines=False)
corpus = data['Please answer the following question. Write between 100 and 150 words. 다음 질문에 100~150단어 답해주십시오.']
```

```python
corpus = corpus.dropna()
corpus
```

```python
total_corpus = " ".join(list(corpus))

FILTER = re.compile("[\^*&\"\"#:;~“”]")
total_corpus = re.sub(FILTER, "", total_corpus)
total_corpus = re.sub("!!", "!", total_corpus)
total_corpus = re.sub("  ", " ", total_corpus)
total_corpus = re.sub("(\.\.)", ".", total_corpus)
total_corpus = re.sub("(\.\.\.)", ".", total_corpus)
total_corpus = re.sub("\n", " ", total_corpus)
total_corpus = re.sub("[()]", ",", total_corpus)
total_corpus = re.sub('[가-힣]','korean',total_corpus)
```

```python
import nltk
from nltk.tokenize import sent_tokenize
```

```python
word_count = len(total_corpus.split(" "))
word_count
```

```python
final_corpus = [i for i in sent_tokenize(total_corpus) if len(i)>5 and i[0] not in ['1','2','3','4','5','6', '-']]
```

```python
final_corpus
```

```python
df = pd.DataFrame(final_corpus)
df
```

```python
df.to_csv("non-native_gachon.txt", header=None, index=None)
```

```python
os.listdir()
```

```python
total_corpus = ""
for folder in os.listdir():
    if folder=='.DS_Store': continue
    move_dir = USE_dir+"/"+folder
    print(move_dir)
    os.chdir(move_dir)
    txt_list = os.listdir()
    for txt in txt_list:
        f = open(txt, 'r', encoding = 'ISO-8859-1')
        data = f.read()
        total_corpus += data
```

```python
regex = "\(.*\)|\s-\s.*"
total_corpus = re.sub(regex, '', total_corpus) 
total_corpus = re.sub("  ", " ", total_corpus)
total_corpus = re.sub("(\.\.)", ".", total_corpus)
total_corpus = re.sub("(\.\.\.)", ".", total_corpus)
total_corpus = re.sub("\n", " ", total_corpus)
```

```python
final_corpus = [i for i in sent_tokenize(total_corpus) if len(i)>5 and i[0] != "<" and not re.search('</title>', i) \
    and not re.search('</doc>', i)]
```

```python
final_corpus = pd.DataFrame(final_corpus)
final_corpus.to_csv('non-native_USE_a-b.txt', header=None, index=None)
```
